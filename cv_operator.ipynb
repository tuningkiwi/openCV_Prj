```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 


img1 = cv2.imread('images/flower1.jpg')
img2 = cv2.imread('images/flower2.jpg')
```

```python
x = np.uint8([30,150,100])
y = np.uint8([230,210,150])
print( cv2.add(x,y) ) # 6250+10 = 20 => 255
#  Saturation연산은 한계값을 정하고 그 값을 벗어나는 경우는 모두 특정 값으로 계산하는 방식 입니다.
# 이미지에서는 0이하는 모두 0, 255이상은 모두 255로 표현하는 것입니다.
print( x+y ) # 250+10 = 260 % 256 = 4
```

```python
dst = cv2.add(img1,img2)
cv2.imshow("dst",dst)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.imshow(dst)
plt.show()
```

```python
#가중치 
# g(x)=(1−α)f0(x)+αf1(x)
# dst=α⋅img1+β⋅img2+γ

dst = cv2.addWeighted(img1,0.7,img2,0.3,0)
cv2.imshow('dst',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.imshow(dst)
plt.show()
```

```python
dst = cv2.addWeighted(img1,0.3,img2,0.7,0)
cv2.imshow('dst',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.imshow(dst)
plt.show()
```
