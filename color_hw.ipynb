```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("./images/airplane1.jpg")
print(img.shape)
print(img)
cv2.imshow("airplane_origin",img)

#img2 = cv2.imread("./images/airplane1.jpg", cv2.IMREAD_GRAYSCALE)
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(img_gray.shape)
print(img_gray)
cv2.imshow("airplane_gray",img_gray)

img_hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
print(img_hsv.shape)
cv2.imshow("airplane_hsv",img_hsv)

img_ycrcb = cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)
print(img_ycrcb.shape)
cv2.imshow("airplane_ycrcb",img_ycrcb)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("./images/airplane1.jpg")
cv2.imshow("airplane_origin",img)
b,g,r =cv2.split(img)
cv2.imshow("b_airplane",b)
cv2.imshow("g_airplane",g)
cv2.imshow("r_airplane",r)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("./images/airplane1.jpg")

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
print(hsv)
hue_val, sat_val, val_val = cv2.split(hsv)

cv2.imshow("src", img)
cv2.imshow("hue",hue_val)
cv2.imshow("saturation",sat_val)
cv2.imshow("value",val_val)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#img = cv2.imread("images\house.jpg")
img = cv2.imread("./images/airplane1.jpg")

img_ycrcb = cv2.cvtColor(img, cv2.COLOR_BGR2YCR_CB)
print(img_ycrcb)
y_luminance, cr_val, cb_val = cv2.split(img_ycrcb)

cv2.imshow("src", img)
cv2.imshow("y_luminance",y_luminance)
cv2.imshow("cr_val",cr_val)
cv2.imshow("cb_val",cb_val)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread("images\house.jpg")

imageHSV = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

imageThresholded = cv2.inRange(imageHSV, (100,0,0), (120,255,255))

cv2.imshow("Thresholded Image", imageThresholded)
cv2.imshow("Original", image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
