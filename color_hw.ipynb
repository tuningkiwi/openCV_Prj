```python
##########################
# BGR SPLIT AND MERGE 
#####################

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("./images/airplane1.jpg")
cv2.imshow("airplane_origin",img)
b,g,r =cv2.split(img)
cv2.imshow("b_airplane",b)
cv2.imshow("g_airplane",g)
cv2.imshow("r_airplane",r)
dst = cv2.merge([b,g,r])
cv2.imshow("dst",dst)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
##########################
# BGR TO HSV
#####################

img = cv2.imread("./images/airplane1.jpg")

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2.imshow("hsv",hsv)

print(hsv)
hue_val, sat_val, val_val = cv2.split(hsv)

cv2.imshow("src", img)

#hsv로 변환된 데이터는 아래와 같이 imshow를 보는 것은 의미없다 
cv2.imshow("hue",hue_val)
cv2.imshow("saturation",sat_val)
cv2.imshow("value",val_val)

print(hue_val.shape, sat_val.shape, val_val.shape)
dst= cv2.merge([hue_val,sat_val,val_val])
cv2.imshow("dst",dst)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
##########################
# BGR TO YCRCB
# 영상의 색감은 유지하고, 명암비를 높이기 
#####################
import cv2


file_list = ["./images/airplane1.jpg", "./images/house.jpg", 
             "./images/baboon.jpg", "./images/flower2.jpg",
             "./images/red_sky.jpg", "./images/puppies.jpg"]


for i, file in enumerate(file_list) :
    img = cv2.imread(file)
    img_ycrcb = cv2.cvtColor(img, cv2.COLOR_BGR2YCR_CB)
    print(img_ycrcb)
    y, cr, cb = cv2.split(img_ycrcb) # y_luminance, cr_val, cb_val 

    y = cv2.equalizeHist(y)
    img_after_eqHist = cv2.merge([y,cr,cb])
    img_final = cv2.cvtColor(img_after_eqHist, cv2.COLOR_YCR_CB2BGR)
    cv2.imshow(f"before{i}",img)
    cv2.imshow(f"after{i}",img_final)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
##########################
# THRESHOLD 및 MASK 사용하기 
#####################

import cv2
import numpy as np

img1 = cv2.imread('images/logo.png')
img2 = cv2.imread('images/lena.jpg')

# 삽입할 이미지의 row, col, channel정보
rows, cols, channels = img1.shape

# 대상 이미지에서 삽입할 이미지의 영역을 추출
roi = img2[0:rows, 0:cols]

#mask를 만들기 위해서 img1을 gray로 변경후 binary image로 전환
#mask는 logo부분이 흰색(255), 바탕은 검은색(0)
#mask_inv는 logo부분이 검은색(0), 바탕은 흰색(255)

img2gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
ret, mask = cv2.threshold(img2gray, 10, 255, cv2.THRESH_BINARY)
mask_inv = cv2.bitwise_not(mask)

#bitwise_and 연산자는 둘다 0이 아닌 경우만 값을 통과 시킴.
#즉 mask가 검정색이 아닌 경우만 통과가 되기때문에 mask영역 이외는 모두 제거됨.
#아래 img1_fg의 경우는 bg가 제거 되고 fg(logo부분)만 남게 됨.
#img2_bg는 roi영역에서 logo부분이 제거되고 bg만 남게 됨.
img1_fg = cv2.bitwise_and(img1, img1, mask=mask)
img2_bg = cv2.bitwise_and(roi, roi, mask=mask_inv)

#2개의 이미지를 합치면 바탕은 제거되고 logo부분만 합쳐짐.
dst = cv2.add(img1_fg, img2_bg)

#합쳐진 이미지를 원본 이미지에 추가.
img2[0:rows, 0:cols] = dst

cv2.imshow('res', img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

#트랙바 생성 lower hue, upper hue 
import cv2
import numpy as np 

def nothing(int):
    pass

img = cv2.imread("images/color_w.png")
img = cv2.resize(img,dsize=(500,500))
img_hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
bg = np.full_like(img,255)
hval, sval, vval = cv2.split(img_hsv)

cv2.namedWindow('colorPicker')
cv2.createTrackbar('lower_hue','colorPicker',30,179,nothing)
cv2.createTrackbar('upper_hue','colorPicker',0,179,nothing)
while True : 
    lo = cv2.getTrackbarPos('lower_hue','colorPicker')
    up = cv2.getTrackbarPos('upper_hue','colorPicker')
    mask = cv2.inRange(hval,lo,up) #lo up 사이에 있으면 흰색, 않으면 검은색 
    mask_inv = cv2.bitwise_not(mask)
    img1 = cv2.bitwise_and(img,img,mask=mask)
    img2 = cv2.bitwise_and(bg,bg,mask=mask_inv)
    cv2.imshow("img1",img1)
    cv2.imshow("img2",img2)
    final_img = cv2.add(img1,img2)
    cv2.imshow("origin",img)
    cv2.imshow("colorPicker",final_img)
    if cv2.waitKey(1) & 0xff ==27:
        break

cv2.destroyAllWindows()
#getHue = cv2.getTrackbarPos('hue','colorPicker')
#이미지 read, convert hsv, h 만 split 해서 inRange 처리 



```
