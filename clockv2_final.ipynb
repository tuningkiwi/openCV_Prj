```python
import cv2
from datetime import datetime
import numpy as np 

while True: 
    #파렛트 준비 
    img = np.full((500,500,3),255,dtype=np.uint8)

    #시계 원 
    c_radius = 200
    cv2.circle(img,(250,250),c_radius,(255,0,0),-1)
    
    #시간 표시 
    num_clock = np.arange(1,13,1,dtype=int)
    for index in num_clock:        
        hours = index
        xh = (c_radius-25)*np.cos(np.pi/180*(30*(hours-3)))
        yh = (c_radius-25)*np.sin(np.pi/180*(30*(hours-3)))
        print(xh,yh)
        #cv2.line(img,(250,250),(250+int(xh),250+int(yh)),(250,100,0),2)
        cv2.putText(img,str(hours),(int(238+xh),int(257+yh)),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),4,1)

    cur_time = str(datetime.today().time())
    print(cur_time)
    cur_hour = int(cur_time[:2])
    cur_min = int(cur_time[3:5])
    cur_sec= int(cur_time[6:8])
    cur_time = f'{cur_hour}:{cur_min}:{cur_sec}'
    cv2.putText(img,cur_time,(10,50),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),2,-1)

    #초침 green
    cur_sec = cur_sec - 15
    xs = c_radius*np.cos(np.pi/180*(6*cur_sec))
    ys = c_radius*np.sin(np.pi/180*(6*cur_sec))
    print(xs,ys)
    cv2.line(img,(250,250),(250+int(xs),250+int(ys)),(0,255,100),2)

    #분침 red
    cur_min_forhour = cur_min
    cur_min = cur_min - 15 #360/60 = 6도씩 . 90도를 움직이려면 15*6 = 90 
    xm = c_radius*2/3*np.cos(np.pi/180*(6*cur_min))
    ym = c_radius*2/3*np.sin(np.pi/180*(6*cur_min))
    print(xm,ym)
    cv2.line(img,(250,250),(250+int(xm),250+int(ym)),(0,100,200),2)

    #시침
    #cur_hour = 0
    cmin_fh = cur_min_forhour*0.5 #분침에 따른 추가 이동 각도
    cur_hour = (cur_hour%12) -3 #360/12 = 30 도씩. 90도를 움직이려면 30*3 = 90 
    print(cur_hour)
    xh = c_radius/3*np.cos(np.pi/180*(30*cur_hour+cmin_fh))
    yh = c_radius/3*np.sin(np.pi/180*(30*cur_hour+cmin_fh))
    print(xh,yh)
    cv2.line(img,(250,250),(250+int(xh),250+int(yh)),(250,100,0),2)
    cv2.imshow("clock",img)
    
    if cv2.waitKey(1000) & 0xFF ==13: #주요 특수키 코드: 27(ESC), 13(ENTER), 9(TAB)
        break
     
cv2.destroyAllWindows()
    
    



```

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
img = cv.imread('gradient.png', cv.IMREAD_GRAYSCALE)
assert img is not None, "file could not be read, check with os.path.exists()"
ret,thresh1 = cv.threshold(img,127,255,cv.THRESH_BINARY)
ret,thresh2 = cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
ret,thresh3 = cv.threshold(img,127,255,cv.THRESH_TRUNC)
ret,thresh4 = cv.threshold(img,127,255,cv.THRESH_TOZERO)
ret,thresh5 = cv.threshold(img,127,255,cv.THRESH_TOZERO_INV)
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
